```python
#Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress
from pathlib import Path
```

```python
energy_file = Path("Resources/Energy_data.csv")
traffic_file = Path("Resources/Traffic_data.csv")
gas_file = Path("Resources/Gas_data.csv")
```

```python
energy_df = pd.read_csv(energy_file)
energy_df.head()
```

```python
energy_df_sort = energy_df.sort_values(by="Current Charges",ascending=False)
energy_df_sort.head()
```

```python
energy_df_clean = energy_df_sort.drop(labels=180730, axis=0)
energy_df_clean.head()
```

```python
energy_df_clean =energy_df_clean.sort_values(by="Revenue Month",ascending=True)
energy_df_clean.head()
```

```python
traffic_df = pd.read_csv(traffic_file)
traffic_df.head()
```

```python
gas_df = pd.read_csv(gas_file)
gas_df.head()
```

```python
#1(A). How has the energy price and energy consumption changed over time? Are they correlated?
# Creating Energy Charges Dataframe
energy_charges = energy_df_clean.groupby("Revenue Month")["Current Charges"].mean()
energy_charges.head()
```

```python
#1(A). How has the energy price and energy consumption changed over time? Are they correlated?
# Creating Energy Consumption Dataframe
energy_consumption = energy_df_clean.groupby("Revenue Month")["Consumption (KWH)"].mean()
energy_consumption.head()
```

```python
#1(A). How has the energy price and energy consumption changed over time? Are they correlated?
# Scatter Plot
plt.scatter(energy_consumption, energy_charges)
plt.xlabel("Energy Consumption by Month")
plt.ylabel("Energy Charges by Month")
plt.title("Correlation Between Average Energy Consumption and Average Energy Price")
plt.xticks(rotation = 90)
corr_coef = energy_consumption.corr(energy_charges)
# print the correlation coefficient
print('Correlation coefficient:', corr_coef)
# calculate the linear regression model
slope, intercept, r_value, p_value, std_err = linregress(energy_consumption, energy_charges)
# print the linear regression equation
print('Linear regression equation: y = {}x + {}'.format(slope, intercept))
# Calculate the linear regression model for Energy Consumption vs. Energy Price
slope, intercept, rvalue, pvalue, stderr = linregress(energy_consumption, energy_charges)
regress_values = energy_consumption * slope + intercept
plt.plot(energy_consumption, regress_values, 'r')
# Display the plot
plt.show()
```

```python
#1(B) How has the energy price and energy consumption changed over time? Are they correlated?
# Line Chart
# Setting Variables
x = energy_consumtion["Revenue Month"]
y = energy_consumption["B&T"]/10000000
z = energy_charges["NY Conventional Gasoline Spot Price ($/gal)"]
# plot lines
plt.plot(x, y, label = "Bridges & Tunnels")
plt.plot(x, z, label = "Gas Price")
plt.legend()
plt.xticks(range(1, 48, 2))
plt.xticks(rotation = 90)
plt.xlabel("Month")
plt.ylabel("B&T (in 10M)")
plt.title("Bridges & Tunnels vs. Gas Price")
plt.show()
```

```python
#2(A) Has the consumer usage of cars/buses affected gas prices?
# Line Chart
# Combine the data into a single DataFrame
combined_data = pd.merge(traffic_df, gas_df, on="Month")
#combined_data.head()
# Line Chart
# Setting Variables
x = combined_data["Month"]
y = combined_data["B&T"]/10000000
z = combined_data["NY Conventional Gasoline Spot Price ($/gal)"]
# plot lines
plt.plot(x, y, label = "Bridges & Tunnels")
plt.plot(x, z, label = "Gas Price")
plt.legend()
plt.xticks(range(1, 48, 2))
plt.xticks(rotation = 90)
plt.xlabel("Month")
plt.ylabel("B&T (in 10M)")
plt.title("Bridges & Tunnels vs. Gas Price")
plt.show()
```

```python
#2(B) Correlation Coefficient and Regression Model
# select the first 24 rows of the dataframe
combined_data_first_24 = combined_data.iloc[:24]
#combined_data_first_24
# calculate the correlation coefficient
b = combined_data_first_24["B&T"]/10000000
g = combined_data_first_24["NY Conventional Gasoline Spot Price ($/gal)"]
corr_coef = b.corr(g)
# print the correlation coefficient
print('Correlation coefficient:', corr_coef)
# Create a scatter plot of B&T vs. Gas Price
plt.scatter(b, g)
# Set the x-axis and y-axis labels and title
plt.xlabel('B&T')
plt.ylabel('Gas Price')
plt.title('B&T vs. Gas Price')
# calculate the linear regression model
slope, intercept, r_value, p_value, std_err = linregress(b, g)
# print the linear regression equation
print('Linear regression equation: y = {}x + {}'.format(slope, intercept))
# Calculate the linear regression model for B&T vs. Gas Price
slope, intercept, rvalue, pvalue, stderr = linregress(b, g)
regress_values = b * slope + intercept
plt.plot(b, regress_values, 'r')
# Display the plot
plt.show()
```

```python
#4(A). Has preference for public transportation changed after 2020?
# Before Pie Chart
before_df = traffic_df[(traffic_df['Month'] < "2020-03")]
total_bus = sum(before_df['Bus Ridership'])
total_subway = sum(before_df['Subway Ridership'])
total_aar = sum(before_df['AAR Ridership'])
total_lirr = sum(before_df['LIRR Ridership'])
total_mnr = sum(before_df['MNR Ridership'])
total_BT = sum(before_df['B&T'])

#Plot a pie chart
labels = ['Bus', 'Subway','Long Island Railroad','Bridges & Tunnels','Metro North Railroad','AAR']
pie_pd = [total_bus, total_subway, total_lirr,total_BT, total_mnr, total_aar]
explode = (0, 0, 0, 0, 0, 0)
fig = plt.figure(figsize = (10,10))
plt.pie(pie_pd, labels = labels, explode = explode, autopct='%1.1f%%')
plt.title("NYC Ridership Preference Before Covid")
plt.show()
```

```python
#4(B). Has preference for public transportation changed after 2020?
# After Pie Chart
#after_df = #create dataframe for before 2020 transportation
after_df = traffic_df[(traffic_df['Month'] >= "2020-03")]
after_clean = after_df.drop(labels=[46,47], axis=0)
total_bus = sum(after_clean['Bus Ridership'])
total_subway = sum(after_clean['Subway Ridership'])
total_aar = sum(after_clean['AAR Ridership'])
total_lirr = sum(after_clean['LIRR Ridership'])
total_mnr = sum(after_clean['MNR Ridership'])
total_BT = sum(after_clean['B&T'])

#plot a pie chart
labels = ['Bus', 'Subway','Long Island Railroad','Bridges & Tunnels','Metro North Railroad','AAR']
pie_pd = [total_bus, total_subway, total_lirr,total_BT,total_mnr, total_aar]
explode = (0, 0, 0, 0, 0, 0)
fig = plt.figure(figsize = (10,10))
plt.pie(pie_pd, labels = labels, explode = explode, autopct='%1.1f%%')
plt.title("NYC Ridership Preference After Covid")
plt.show()
```
