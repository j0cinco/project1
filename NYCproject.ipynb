```python
#Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress
from pathlib import Path
```

```python
energy_file = Path("Resources/Energy_data.csv")
traffic_file = Path("Resources/Traffic_data.csv")
gas_file = Path("Resources/Gas_data.csv")
```

```python
energy_df = pd.read_csv(energy_file)
energy_df.head()
```

```python
traffic_df = pd.read_csv(traffic_file)
traffic_df.head()
```

```python
gas_df = pd.read_csv(gas_file)
gas_df.head()
```

```python
#1(A). How has the energy price and energy consumption changed over time? Are they correlated?
# Histogram
# Energy Price Over Time
fig,ax = plt.subplots(1,1)
a = energy_df["Current Charges"]
ax.hist(a, bins = [len(energy_df["Revenue Month"])])
ax.set_title("Energy Price Over TIme Before and After Covid")
ax.set_xlabel('Date by Month')
ax.set_ylabel('Energy Price')
plt.show()
```

```python
#1(A). How has the energy price and energy consumption changed over time? Are they correlated?
# Histogram
# Energy Consumption (KWH) Over Time
fig,ax = plt.subplots(1,1)
a = energy_df["Consumption (KWH)"]
ax.hist(a, bins = [len(energy_df["Revenue Month"])])
ax.set_title("Energy Price Over TIme Before and After Covid")
ax.set_xlabel('Date by Month')
ax.set_ylabel('Energy Consumption (KWH)')
plt.show()
```

```python
#1(B). How has the energy price and energy consumption changed over time? Are they correlated?
# Line Chart
# Energy Price Over Time
plt.scatter(energy_df["Revenue Month"], energy_df["Current Charges"])
plt.xlabel("Date by Month")
plt.ylabel("Energy Price")
plt.title("NYC Price of Energy Before and After Covid")
plt.xticks(rotation = 90)
plt.show()
```

```python
#1(B). How has the energy price and energy consumption changed over time? Are they correlated?
# Line Chart
# Energy Consumption Over Time
plt.scatter(energy_df["Revenue Month"], energy_df["Consumption (KWH)"])
plt.xlabel("Date by Month")
plt.ylabel("Energy Consumption (KWH)")
plt.title("NYC Consumption of Energy Before and After Covid")
plt.xticks(rotation = 90)
plt.show()
```

```python
#3(A). Has public transportation usage returned to pre-pandemic levels?
# Line Chart
# Setting Variables
x = traffic_df["Month"]
y = traffic_df["Bus Ridership"]
z = traffic_df["Subway Ridership"]
a = traffic_df["AAR Ridership"]
b = traffic_df["LIRR Ridership"]
c = traffic_df["MNR Ridership"]
d = traffic_df["B&T"]
# plot lines
plt.plot(x, y, label = "Bus Ridership")
plt.plot(x, z, label = "Subway Ridership")
plt.plot(x, a, label = "AAR Ridership")
plt.plot(x, b, label = "LIRR Ridership")
plt.plot(x, c, label = "MNR Ridership")
plt.plot(x, d, label = "B&T")
plt.legend()
plt.xticks(rotation = 90)
plt.xlabel("Date by Month")
plt.ylabel("Ridership Number")
plt.title("NYC Ridership by Type Before & After Covid")
plt.show()
```

```python
#3(B). Has public transportation usage returned to pre-pandemic levels?
# Bar Chart
# Setting Variables
x = traffic_df["Month"]
y = traffic_df["Bus Ridership"]
z = traffic_df["Subway Ridership"]
a = traffic_df["AAR Ridership"]
b = traffic_df["LIRR Ridership"]
c = traffic_df["MNR Ridership"]
d = traffic_df["B&T"]
# plot bars
plt.bar(x, y, label = "Bus Ridership")
plt.bar(x, z, label = "Subway Ridership")
plt.bar(x, a, label = "AAR Ridership")
plt.bar(x, b, label = "LIRR Ridership")
plt.bar(x, c, label = "MNR Ridership")
plt.bar(x, d, label = "B&T")
plt.legend()
plt.xticks(rotation = 90)
plt.xlabel("Date by Month")
plt.ylabel("Ridership Number")
plt.title("NYC Ridership by Type Before & After Covid")
plt.show()
```
